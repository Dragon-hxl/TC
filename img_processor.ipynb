```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
#image size 596*1264

imgpath = ["E:/Training Set/TrainingSet1","E:/Training Set/TrainingSet2","E:/Training Set/TrainingSet3","E:/Training Set/TrainingSet4",
           "E:/Training Set/TrainingSet5","E:/Training Set/TrainingSet6","E:/Training Set/TrainingSet7","E:/Training Set/TrainingSet8"]
savepath = "E:/Means/Means3"
images_beforeL = np.zeros(596*1264)
images_afterL = np.zeros(596*1264)
images_beforeR = np.zeros(596*1264)
images_afterR = np.zeros(596*1264)

a = 72
d = 24

def y(x):
    y = -a*(x-d/2)*(x+d/2)
    if y < 0:
        y = 0
    return y

def get_means_r2(images):
    num = len(images) 
    mid = num/2
    vsum = 0
    for i in range(num):#range 默认从0开始
        vsum += y(i-mid)
    factor = 1.0/vsum
    means = np.zeros(images[0].shape)
    for i in range(num):
        means += images[i]*y(i-mid)*factor
    return means

def get_means_evenly(images):
    num = len(images)
    means = np.zeros(images[0].shape)
    for i in range(num):
        means += images[i]
    means = means/num
    return means

def get_means(images):
    num = len(images)
    if num <= 6:
        mean = get_means_evenly(images)
    else:
        mean = get_means_r2(images)
    return mean

if __name__ == "__main__":
    for i in range(len(imgpath)):
        dirs = os.listdir(imgpath[i])
        for _dir in dirs:
            dirp = imgpath[i]+"/"+_dir
            imgs = os.listdir(dirp)
            mean_bL = np.zeros(596*1264)
            mean_bR = np.zeros(596*1264)
            mean_aL = np.zeros(596*1264)
            mean_aR = np.zeros(596*1264)
            for img in imgs:
                if os.path.splitext(img)[1] != ".jpg":
                    continue
                img_p = dirp+"/"+img
                image = cv2.imread(img_p,0)
                if image.shape !=(596,1264):
                    continue
                image=image.reshape(596*1264)
                if img[9]=="L" and img[11]=="1":
                    mean_bL = np.vstack((mean_bL,image))
                    #mean_bL[0] += image 
                if img[9]=="L" and img[11]=="2":
                    mean_aL = np.vstack((mean_aL,image))
                    #mean_aL[0] += image
                if img[9]=="R" and img[11]=="1":
                    mean_bR = np.vstack((mean_bR,image))
                    #mean_bR[0] += image
                if img[9]=="R" and img[11]=="2":
                    mean_aR = np.vstack((mean_aR,image))
                    #mean_aR[0] += image
            if np.ndim(mean_bL) > 1:
                meanbl = get_means(mean_bL[1:]).reshape((596,1264))
                cv2.imwrite(savepath+"/"+_dir+"BLM.jpg",meanbl)
            if np.ndim(mean_aL) > 1:
                meanal = get_means(mean_aL[1:]).reshape((596,1264))
                cv2.imwrite(savepath+"/"+_dir+"ALM.jpg",meanal)
            if np.ndim(mean_bR) > 1:
                meanbr = get_means(mean_bR[1:]).reshape((596,1264))
                cv2.imwrite(savepath+"/"+_dir+"BRM.jpg",meanbr)
            if np.ndim(mean_aR) > 1:
                meanar = get_means(mean_aR[1:]).reshape((596,1264))
                cv2.imwrite(savepath+"/"+_dir+"ARM.jpg",meanar)
            #print("Finish!"+_dir)
```
